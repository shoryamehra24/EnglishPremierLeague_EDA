# EDA of Premier League 2020-21

# Player Statistics

```python
most_goals = epl_df.nlargest(10,"Goals").reset_index()
most_goals[["Name","Goals"]].style.background_gradient(cmap="flare")
```

```python
sns.set(rc={'figure.figsize':(20,7.5)})
plt.title("Most Goals in th Premier League", fontsize=20)
sns.barplot(x="Name",y="Goals",data=most_goals, palette='flare')
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
plt.show()
```

```python
m_assists = epl_df.nlargest(10, "Assists")
m_assists[['Name', "Assists"]].reset_index().drop('index',axis=1).style.background_gradient(cmap="flare")
```

```python
sns.barplot(x="Name",y="Assists",data=m_assists, palette='flare')
plt.title("Most Assists in the Premier League", fontsize=20)
locs,labels = plt.xticks()
plt.setp(labels, rotation=45)
plt.show()
```

```python
m_passes = epl_df.nlargest(10,"Passes_Attempted")
m_passes[["Name","Passes_Attempted"]].reset_index().drop('index',axis=1).style.background_gradient(cmap='flare')
```

```python
sns.barplot(x="Name",y="Passes_Attempted",data=m_passes, palette='flare')
plt.title("Most Passes Attempted in the Premier League", fontsize=20,)
plt.ylabel("Passes Attempted")
locs,labels = plt.xticks()
plt.setp(labels, rotation=45)
plt.show()
```

```python
m_penalties = epl_df.nlargest(10, "Penalty_Goals")
m_penalties[["Name","Penalty_Goals"]].reset_index().drop('index',axis=1).style.background_gradient(cmap="flare")
```

```python
sns.barplot("Name","Penalty_Goals",data=m_penalties, palette='flare')
plt.title("Most Penalty Goals in the Premier League", fontsize=20)
plt.ylabel("Penalty Goals")
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
plt.show()
```

```python
m_yellow = epl_df.nlargest(10,"Yellow_Cards")
m_yellow[["Name","Yellow_Cards"]].reset_index().drop('index',axis=1).style.background_gradient(cmap="flare")
```

```python
sns.barplot(x="Name",y="Yellow_Cards",data=m_yellow, palette='flare')
plt.ylabel("Yellow Cards", fontsize=20)
plt.title("Most Yellow Cards In the Premier League")
plt.xticks(rotation=45, style='oblique')
plt.show()
```

```python
m_red = epl_df.nlargest(10,"Red_Cards")
m_red[["Name","Red_Cards"]].reset_index().drop('index',axis=1).style.background_gradient(cmap="flare")
```

```python
sns.barplot(x="Name",y="Red_Cards",data=m_red, palette='flare')
plt.ylabel("Red Cards")
plt.title("Most Red Cards In the Premier League", fontsize=20)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
plt.show()
```

# Club Statistics

```python
data = ["Goals","Assists","Passes_Attempted", 'Penalty_Goals', 'Yellow_Cards',
       'Red_Cards']

for column in data:
    team = pd.DataFrame(epl_df.groupby("Club")[column].sum().reset_index(drop=False).sort_values(by=column,ascending=False))
    display(team.style.background_gradient(cmap="crest"))
    sns.barplot(x="Club",y=column,data= team, palette='crest')
    plt.title(f"The most {column} in the Premiere League", fontsize=20)
    plt.xlabel(' ')
    plt.ylabel(f'count of {column}',fontsize=18)
    plt.xlabel(' ')
    plt.xticks(ticks=np.arange(0,20,1),labels=team['Club'],fontsize=18,style='oblique',rotation=90)
    plt.yticks(fontsize=18)
    plt.show()
    
```
